```python
import firebase_admin
from firebase_admin import credentials, firestore
import random
import time

# Path to your Firebase service account key JSON file
service_account_path = "kunci_firebase.json"

# Initialize the Firebase app with the service account
cred = credentials.Certificate(service_account_path)
firebase_admin.initialize_app(cred)

# Initialize Firestore
db = firestore.client()

# Fungsi untuk menghitung titik embun
def hitung_titik_embun(suhu, kelembapan):
    a = 17.27
    b = 237.7
    alpha = ((a * suhu) / (b + suhu)) + (kelembapan / 100.0)
    titik_embun = (b * alpha) / (a - alpha)
    return round(titik_embun, 2)

# Fungsi untuk menghasilkan data acak dan mengirimnya ke Firestore
def kirim_data_logger(sensor_id):
    try:
        # Menghasilkan data acak
        suhu = round(random.uniform(20, 35), 2)  # Suhu antara 20-35Â°C
        kelembapan = round(random.uniform(40, 80), 2)  # Kelembapan antara 40-80%
        tekanan = round(random.uniform(990, 1030), 2)  # Tekanan udara antara 990-1030 hPa
        titik_embun = hitung_titik_embun(suhu, kelembapan)  # Hitung titik embun
        
        # Unix timestamp
        timestamp = int(time.time())

        # Data yang akan dikirimkan
        data = {
            "suhu": suhu,
            "kelembapan": kelembapan,
            "tekanan_udara": tekanan,
            "titik_embun": titik_embun,
            "timestamp": timestamp
        }

        # Menyimpan data di Firestore
        # Koleksi 'sensor_iklim', subkoleksi berdasarkan sensor_id (misal 'id-01'), dokumen dengan timestamp
        db.collection("sensor_iklim").document(sensor_id).collection("data").document(str(timestamp)).set(data)
        
        print(f"Data berhasil dikirim ke sensor_iklim/{sensor_id}/data/{timestamp}: {data}")
    except Exception as e:
        print(f"Gagal mengirim data: {e}")

# Contoh penggunaan
sensor_id = "id-01"  # ID sensor, bisa diubah sesuai sensor yang digunakan

while True:
    # Mengirim data ke subkoleksi 'data' dari sensor_id setiap 60 detik
    kirim_data_logger(sensor_id)
    
    # Tunggu 60 detik sebelum mengirim data berikutnya
    time.sleep(10)
```
